## COUNTING

```python
def unique_words(s):
    unique=set(s.split(" "))
    return len(unique)/50

def emoji(posts):
    emoji_count=0
    words=posts.split()
    for i in words:
        if "http" not in i:
            if i.count(':')==2:
                emoji_count+=1
    return emoji_count/50

def colon(posts):
     # Includes colons used in emojis
    colon_count=0
    words=posts.split()
    for i in words:
        if "http" not in i:
            colon_count+=i.count(":")
    return colon_count/50
```

```python
# Counting question mark,exlamatory mark , emojis,colons

df["qm"] = df["posts"].apply(lambda x: x.count("?") / 50)
df['em'] = df['posts'].apply(lambda x:x.count("!") /50)
df['emojis']=df['posts'].apply(emoji)
df['colon']=df['posts'].apply(colon)
```

```python
# Counting words and unique words
df['unique_words']=df['posts'].apply(unique_words)
df['word_count']=df['posts'].apply(lambda x:(x.count(" ")+1)/50)
```

```python
# Upper case count
df['upper_case']=df['posts'].apply(lambda x:len([x for x in x.split(" ") if x.isupper()])/50)
```

```python
# link count

df['link_count']=df['posts'].apply(lambda x:x.count('http')/50)
```

```python
# Ellipses count
df['ellipses_count']=[
    len(re.findall(r"\.\.\.\ ", posts)) / 50 for posts in df["posts"]
]
```

```python
# Image count
df['img_count']=[
    len(re.findall(r"(\.jpg)|(\.jpeg)|(\.gif)|(\.png)", posts)) / 50 for posts in df["posts"]
]
```

```python
df['post_length']=df['posts'].apply(lambda x:np.var([len(post.split()) for post in x.split("|||")]))
```

```python
df.to_csv("D:\\data\\clean_data_3.csv",index=False)
```
